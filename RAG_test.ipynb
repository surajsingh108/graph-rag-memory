```python

# 3. Add data to the collection
document = "The history of Rome is fascinating. It started with..."
collection.add(
    documents=[document],
    ids=["rome_history_doc"]
)
```

```python

# 4. Construct the Prompt
prompt = "Tell me more about the history of Rome."

# 5. Generate Text with Augmented Prompt
print("\nGenerating text...")
generator = pipeline('text-generation', model='gemma-2b', device=0 if torch.cuda.is_available() else -1)
generated_text = generator(prompt, max_length=100, do_sample=True, temperature=0.7)

# 7. Print the generated text
print("\nGenerated Text:")
print(generated_text[0]['generated_text'])
```
